### 도로명 주소 구하기

```python
import requests
from urllib.parse import urlparse
import pandas as pd
```

```python
key_fd = open('roadapikey.txt', mode='r')
roadapi_key = key_fd.read(100)
key_fd.close()
```

```python
df = pd.read_csv('커피업소정보.csv',encoding='euc-kr')
df.head()
```

```python
tmp = []
tmp2 = []
for e in df['도로명주소'].tolist():
    tmp2.append(e.split()[0])
    tmp.append(e.split()[1])
df['tmp2'] = tmp2
df['tmp'] = tmp
df.head()
```

```python
df2 = df[df.tmp.isin(['강서구'])&df.tmp2.isin(['서울특별시'])]
df2 = df2.reset_index()
```

```python
del df2['index']
```

```python
df2.head()
```

```python
key_fd = open('openweatherkey.txt', mode='r')
oweather_key = key_fd.read(100)
key_fd.close()
```

```python
from tqdm import tqdm_notebook
```

```python
weather_list = []; desc_list = []; temp_list = []
for i in tqdm_notebook(df2.index):
    lat = df2['위도'][i]
    lng = df2['경도'][i]
    url = f'http://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lng}&appid={oweather_key}&units=metric&lang=kr'
    result = requests.get(urlparse(url).geturl()).json()
    main = result['weather'][0]['main']
    desc = result['weather'][0]['description']
    temp = round(result['main']['temp']-0.01,1)
    temp_min = round(result['main']['temp_min']-0.01,1)
    temp_max = round(result['main']['temp_max']-0.01,1)
    weather_list.append(main)
    desc_list.append(desc)
    temp_list.append(f'{temp}/{temp_min}/{temp_max}')
```

```python
df2['weather'] = weather_list
df2['desc'] = desc_list
df2['temp'] = temp_list
df2
```

```python
df2.to_csv('cafe_kangseo.csv',sep=',',encoding='utf8')
```

```python
df2['위도'].mean(),df2['경도'].mean()
```

```python
color_dict = {'Clear' : 'blue', 'Snow': 'white', 'Rain':'gray','Extreme':'red','Clouds':'orange','Mist':'green'}
```

```python
import folium
```

```python
mapping = folium.Map(location=[df2['위도'].mean(),df2['경도'].mean()], zoom_start=12)
mapping
```

```python
for i in df2.index:
    folium.Marker(
        location=[df2['위도'][i],df2['경도'][i]], 
        popup= df2['상호명'][i],
        tooltip = f'{df2.desc[i]}, {df2.temp[i]}',
        icon=folium.Icon(color = color_dict[df2.weather[i]])
    ).add_to(mapping)

mapping
```

```python
name_list = df2['상호명'].tolist()
```

```python
e_list = [];s_list = [];p_list = [];t_list = []
for name in name_list:
    if '이디야' in name:
        e_list.append(name)
    elif '스타벅스' in name:
        s_list.append(name)
    elif '빽다방' in name:
        p_list.append(name)
    elif '투썸플레이스' in name:
        t_list.append(name)
```

```python
df2_e = df2[df2['상호명'].isin(e_list)]
df2_s = df2[df2['상호명'].isin(s_list)]
df2_p = df2[df2['상호명'].isin(p_list)]
df2_t = df2[df2['상호명'].isin(t_list)]
```

```python
df3 = pd.concat([df2_e,df2_s,df2_p,df2_t])
```

```python
df3 = df3.reset_index()
```

```python
del df3['index']
```
