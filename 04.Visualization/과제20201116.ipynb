```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
mpl.rc('font', family='Malgun Gothic')
mpl.rc('axes', unicode_minus=False)
```

```python
df_cafe = pd.read_csv('cafe_2year.csv')
df_cafe
```

```python
plt.figure(figsize=(20,5))
sns.barplot(data=df_cafe, x='기준월', y='카페수')
plt.show()
```

```python
df_cafe_3 = df_cafe[df_cafe.index%3 == 2]
```

```python
sns.barplot(data=df_cafe_3, x='기준월', y='카페수', palette="Blues_d")
plt.ylim(df_cafe_3["카페수"][2],df_cafe_3["카페수"][23])
plt.yticks(df_cafe_3['카페수'])
plt.xlabel('년월')
plt.ylabel('카페수')
plt.show()
```

```python
# 연습문제 5
df_doc2 = pd.read_csv('doctor2.csv')
df_doc2
```

```python
x = df_doc2['지역']
a = df_doc2['일반의 수']
b = df_doc2['인턴 수']
c = df_doc2['레지던트 수']
d = df_doc2['전문의 수']

plt.plot(x ,a, '-',color = 'orange', marker = 'o',label = '일반의')
plt.plot(x ,b, '--',marker = 'x',label = '인턴')
plt.plot(x ,c, 'y',marker = 's',label = '레지턴트')
plt.plot(x ,d,'-.',color = 'purple', marker = 'D',label = '전문의')
plt.legend()
plt.show()
```

```python
# 연습문제 6
df_seoul = df_doc2.iloc[0].tolist()
df_seoul
```

```python
ratio = df_seoul[1:]
explode = [0,0.05,0,0]
labels = ['일반의','인턴','레지던트','전문의']
plt.pie(ratio, labels=labels, autopct='%.1f%%',startangle=90,explode=explode)
plt.title('서울 종합병원 의사 유형별 분포')
plt.show()
```
