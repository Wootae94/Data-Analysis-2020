#### 08. Matplotlib 그래프 영역 채우기

```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_between(x[1:3], y[1:3], alpha = 0.8)
plt.axhline(0); plt.axvline(0)
plt.show()
```

```python
plt.plot(x, y)
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.fill_betweenx(y[2:4],x[2:4],color = 'pink', alpha = 0.5)
plt.axhline(0); plt.axvline(0)
plt.show()
```

```python
x = np.arange(1.0, 4.0,0.01)
y1 = x ** 2
y2 = x ** 2 / 2

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X-Label'); plt.ylabel('Y-Label')
plt.title('두 그래프 사이 영역 채우기')
plt.fill_between(x[100:201], y1[100:201], y2[100:201], color='lightgray', alpha=0.5)

plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [1, 2, 4, 8]

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('X-Label');plt.ylabel('Y-Label')
plt.title('임의의 영역 채우기')
plt.fill([1.9, 1.9, 3.1, 3.1], [2, 5, 11, 8], color='lightgray', alpha=0.5)

plt.show()
```

#### 12. Matplotlib 타이틀 설정하기

#### 13. Matplotlib 수직선/수평선 표시하기

```python
a = np.arange(0, 2, 0.2)

plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.xticks(np.arange(0, 2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

plt.axhline(1, 0, 0.55, color='gray', linestyle='--', linewidth='1')
plt.axvline(1, 0, 0.16, color='lightgray', linestyle=':', linewidth='2')

plt.axhline(5.83, 0, 0.95, color='gray', linestyle='--', linewidth='1')
plt.axvline(1.8, 0, 0.95, color='lightgray', linestyle=':', linewidth='2')

plt.show()
```

```python
plt.plot(a, a, 'bo')
plt.plot(a, a**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(a, a**3, color='springgreen', marker='^', markersize=9)
plt.xticks(np.arange(0, 2, 0.2), labels=['Jan', '', 'Feb', '', 'Mar', '', 'May', '', 'June', '', 'July'])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

plt.hlines(4, 1, 1.6, colors='pink', linewidth=3)
plt.vlines(1, 1, 4, colors='pink', linewidth=3)

plt.show()
```

```python
# sin, cos 그래프 그리기
x = np.arange(0,2 * np.pi,0.02)
plt.plot(x, np.sin(x),label = 'sine')
plt.plot(x, np.cos(x),label = 'cosine')
plt.axis([0,2*np.pi,-1,1])
plt.xticks(np.arange(0,2*np.pi+0.01,np.pi/2),labels=['0',r'$\pi$/2',r'$\pi$',r'3$\pi/2$',r'2$\pi$'])
plt.yticks([-1,0,1],labels=['Low',"Zero",'High'])
plt.grid(True);plt.title('Sine,Cosine 그래프')
plt.legend(loc =3)
plt.show()
```

#### 14. Matplotlib 막대 그래프 그리기

```python
x = np.arange(3)
years = ['2017','2018','2019']
values = [100, 400, 900]

plt.bar(x, values)
plt.xticks(x, years)
plt.show()
```

```python
plt.bar(x, values, width= 0.6, align= 'edge',color = "springgreen",
edgecolor = 'gray', linewidth = 3,tick_label=years, log=True)
plt.title('스타일꾸미기')
plt.show()
```

#### 15. Matplotlib 수평 막대 그래프 그리기

```python
y = np.arange(3)
years = ['2017', '2018', '2019']
values = [100, 400, 900]

plt.barh(y, values, height=-0.6, align='edge', color="springgreen",
        edgecolor="gray", linewidth=3, tick_label=years, log=False)
plt.show()
```

#### 16. Matplotlib 산점도 그리기

```python
np.random.seed(2020)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
plt.scatter(x, y)
plt.show()
```

```python
np.random.seed(2020)    # Reproducible random state

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')
np.random.seed(2020)
x = iris[iris.species == 'setosa'].sepal_length
y = iris[iris.species == 'setosa'].sepal_width
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2
plt.title('Iris Setosa 품종')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal width')
plt.scatter(x, y, s=area, c=colors,alpha=0.5)
plt.show()
```

#### 18. Matplotlib 히스토그램 그리기

```python
weight = [68, 81, 64, 56, 78, 74, 61, 77, 66, 68, 59, 71, 80, 59, 67, 81, 69, 73, 69, 74, 70, 65]

plt.hist(weight)
plt.show()
```

```python
np.random.seed(2020)
x = np.random.randn(1000)
plt.hist(x)
plt.show()
```

```python
np.random.seed(2020)
x = np.random.randn(100000)
plt.hist(x, bins = 100)
plt.show()
```

```python
# 표준정규분포 그래프
x = np.linspace(-4.0,4.0,801)
nd = np.exp(-x*x/2)/np.sqrt(2*np.pi)
plt.plot(x, nd)
plt.axhline(0, alpha = 0.5);plt.axvline(0, alpha = 0.5)
plt.show()
```

```python
a = 2.0 * np.random.randn(10000) + 1.0
b = np.random.standard_normal(10000)
c = 20.0 * np.random.rand(5000) - 10.0

plt.hist(a, bins=100, density=True, alpha=0.7, histtype='step',label= 'a')
plt.hist(b, bins=50, density=True, alpha=0.5, histtype='stepfilled',label= 'b')
plt.hist(c, bins=100, density=True, alpha=0.9, histtype='step',label= 'c')
plt.legend()
plt.show()
```

#### 20. Matplotlib 파이 차트 그리기

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']

plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.show()
```

```python
ratio = [34, 32, 16, 18]
labels = ['Apple', 'Banana', 'Melon', 'Grapes']
explode = [0, 0.10, 0, 0.10]

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, explode=explode)
plt.show()
```

#### Boxplot

```python
iris = sns.load_dataset('iris')
x1 = iris[iris.species == 'setosa'].sepal_length
y1 = iris[iris.species == 'setosa'].sepal_width
a1 = iris[iris.species == 'setosa'].petal_length
b1 = iris[iris.species == 'setosa'].petal_width
plt.title('Setosa')
plt.boxplot([x1, y1, a1, b1],labels=["Sepal length","Sepal width","Petal length","Petal width"])
plt.show()
```
