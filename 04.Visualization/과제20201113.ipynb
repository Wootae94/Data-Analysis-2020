```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mpl.rc('font', family='Malgun Gothic')
mpl.rc('axes', unicode_minus=False)
```

```python
# 코로나 확진자 데이터
covid19_date = ['04/25','04/26','04/27','04/28','04/29','04/30','05/01']
covid19_confirmed = [10, 10, 10, 14, 9, 4, 4]
plt.plot(covid19_date,covid19_confirmed)
plt.xlabel('날짜')
plt.ylabel('일일 확진자 수')
plt.grid(True)
plt.show()
```

```python
covid19_date = ['04/25','04/26','04/27','04/28','04/29','04/30','05/01']
covid19_acc = [10718,10728,10738,10752,10761,10765,10774]  # 누적확진자 수
covid19_rel = [8635,8717,8764,8854,8922,9059,9072] # 누적 격리해제자 수
plt.plot(covid19_date,covid19_acc,label = '누적확진자 수')
plt.plot(covid19_date,covid19_rel, label = '누적 격리 해제자 수')
plt.xlabel('날짜')
plt.ylabel('인원 수')
plt.legend()
plt.grid(True)
plt.show()
```

```python
y = np.arange(9)
age = ['0-9','10-19','20-29','30-39','40-49','50-59','60-69','70-79','80세 이상']
covid19_confirmed_age = [140,571,2962,1163,1432,1956,1351,710,488]
plt.barh(y, covid19_confirmed_age, height=-0.6, align='edge', color="springgreen",
        edgecolor="gray", linewidth=3, tick_label=age, log=False)
plt.xlabel('확진자 수')
plt.ylabel('연령대')
plt.title('연령대별 확진자 수')
plt.show()
```

```python
x = np.arange(len(age))
age = ['0-9','10-19','20-29','30-39','40-49','50-59','60-69','70-79','80세 이상']
covid19_confirmed_age = [140,571,2962,1163,1432,1956,1351,710,488]
covid19_dead = [0, 0, 0, 2, 3, 15, 35,75,120]
plt.bar(x-0.0, covid19_confirmed_age, align='edge', color="springgreen",
        edgecolor="gray", width=0.2, tick_label=age, label = '확진자 수',log=False)
plt.bar(x+0.2, covid19_dead, align='edge',
        edgecolor="gray", width=0.2, tick_label=age, label = '사망자수', log=False)
plt.legend()
plt.xlabel('인원 수')
plt.ylabel('연령대')
plt.title('연령대별 확진자/사망자 수')
plt.show()
```

```python
covid19_confirmed_age = [140,571,2962,1163,1432,1956,1351,710,488]
age = ['0-9','10-19','20-29','30-39','40-49','50-59','60-69','70-79','80세 이상']
plt.pie(covid19_confirmed_age, labels=age, autopct='%.1f%%', counterclock = False)
plt.title('연령대별 확진 비율')
plt.show()
```
