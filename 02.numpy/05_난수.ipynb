### 난수 생성

```python
np.random.randn(10)
```

```python
np.random.randint(10, size = 10)
```

```python
today = 10000
tommorow = today * (1 + np.random.randn(1)/100.0)
tommorow
```

```python
prices = []
price = 10000
for i in range(250):
    tmp = price * (1 + np.random.randn(1)/100.0)
    price = tmp[0]
    prices.append(price)
prices
```
