```python
import pandas as pd
import time 
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome('./chromedriver')
driver.maximize_window()
driver.get('http://www.opinet.co.kr/searRgSelect.do')
time.sleep(1)
driver.find_element_by_css_selector('.ic_m1').click()
time.sleep(2)
region = driver.find_element_by_xpath('//*[@id="SIGUNGU_NM0"]')
gu_list = region.find_elements_by_tag_name('option')
gu_names = [gu.get_attribute('value') for gu in gu_list]
del gu_names[0]
```

```python
# 25개 자치구에 대해서 엑셀 다운로드
for gu in gu_names:
    region = driver.find_element_by_id("SIGUNGU_NM0")
    region.send_keys(gu)
    time.sleep(1)

    driver.find_element_by_xpath('//*[@id="glopopd_excel"]/span').click()
    time.sleep(2)
```

### 구별 휘발유 가격에 대한 데이터 정리

```python
from glob import glob 
```

```python
glob('./station_data/지역*.xls')
```

```python
df = pd.read_excel('./station_data/지역_위치별(주유소).xls',header=2)
df.head()
```

```python
station_files = glob('./station_data/지역*.xls')
tmp_raw = []

for file_name in station_files:
    tmp = pd.read_excel(file_name, header=2)
    tmp_raw.append(tmp)

station_raw = pd.concat(tmp_raw)
```

```python
station_raw.info()
```

```python
stations = pd.DataFrame({
    '상호' : station_raw['상호'],
    '주소' : station_raw['주소'],
    '가격' : station_raw['휘발유'],
    '셀프여부' : station_raw['셀프여부'],
    '상표' : station_raw['상표']
})
stations.head()
```

```python
stations['구'] = [addr.split()[1] for addr in stations['주소']]
stations.tail()
```

```python
stations['구'].unique(),len(stations['구'].unique())
```

```python
stations = stations[stations['가격'] != '-']
stations.info()
```

```python
stations['가격'] = stations['가격'].astype(int)
stations.info()
```

```python
oil_price_top10 = stations.sort_values(by='가격',ascending=False).head(10).reset_index()
oil_price_bottom10 = stations.sort_values(by='가격',ascending=True).head(10).reset_index()
del oil_price_top10['index']
del oil_price_bottom10['index']
```

```python
del oil_price_top10['index']
del oil_price_bottom10['index']
```

```python
import requests
from urllib.parse import urlparse,quote
import numpy as np
```

```python
key_fd = open('../06.RestAPI/kakaomapkey.txt', mode='r')
kmap_key = key_fd.read(100)
key_fd.close()
```

```python
from tqdm import tqdm_notebook

lat_list = []; lng_list = []
for i in tqdm_notebook(oil_price_top10.index):
    try:
        addr = str(oil_price_top10['주소'][i])
        url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + quote(addr)
        result = requests.get(url, headers = {"Authorization": "KakaoAK " + kmap_key}).json()
        lat = float(result['documents'][0]['y'])
        lng = float(result['documents'][0]['x'])
        lat_list.append(lat)
        lng_list.append(lng)
    except:
        lat_list.append(np.nan)
        lng_list.append(np.nan)

oil_price_top10['lat'] = lat_list
oil_price_top10['lng'] = lng_list
```

```python
oil_price_top10
```

```python
lat_list = []; lng_list = []
for i in tqdm_notebook(oil_price_bottom10.index):
    try:
        addr = str(oil_price_bottom10['주소'][i])
        url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + quote(addr)
        result = requests.get(url, headers = {"Authorization": "KakaoAK " + kmap_key}).json()
        lat = float(result['documents'][0]['y'])
        lng = float(result['documents'][0]['x'])
        lat_list.append(lat)
        lng_list.append(lng)
    except:
        lat_list.append(np.nan)
        lng_list.append(np.nan)

oil_price_bottom10['lat'] = lat_list
oil_price_bottom10['lng'] = lng_list
```

```python
oil_price_bottom10
```

```python
map = folium.Map(location=[37.5202, 126.975], zoom_start=10.5)

for n in oil_price_top10.index:
    if oil_price_top10['lat'][n].any():
        folium.CircleMarker([oil_price_top10['lat'][n], oil_price_top10['lng'][n]], 
                                  radius=15, color='#CD3181', 
                                  fill_color='#CD3181',
                                  fill=True).add_to(map)
    
for n in oil_price_bottom10.index:
    if oil_price_bottom10['lat'][n].any(): 
        folium.CircleMarker([oil_price_bottom10['lat'][n], 
                                  oil_price_bottom10['lng'][n]], 
                                  radius=15, color='#3186cc', 
                                  fill_color='#3186cc',
                                  fill=True).add_to(map)
        
map
```
