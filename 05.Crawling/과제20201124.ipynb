```python
df = pd.read_excel('./station/영등포구 주유소.xls')
df.head()
```

```python
# 컬럼명 재설정
df.columns = df.iloc[1]
df.head()
```

### 도로명 주소로부터 위도, 경도 찾기

```python
import requests
from urllib.parse import urlparse,quote
```

```python
import numpy as np
```

```python
key_fd = open('../06.RestAPI/kakaomapkey.txt', mode='r')
kmap_key = key_fd.read(100)
key_fd.close()
```

```python
lat_list = []; lng_list = []
for i in df.index:
    addr = df['주소'][i]
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + addr
    result = requests.get(urlparse(url).geturl(), headers = {"Authorization": "KakaoAK " + kmap_key}).json()
    try:
        lat = float(result['documents'][0]['y'])
        lng = float(result['documents'][0]['x'])
        lat_list.append(lat)
        lng_list.append(lng)
    except:
        lat = np.nan
        lng = np.nan
        lat_list.append(lat)
        lng_list.append(lng)
```

```python
df['lat'] = lat_list
df['lng'] = lng_list
df
```

```python
df = df.dropna()
```

```python
# 엑셀 다운 안 하고 긁어오기
driver = webdriver.Chrome('./chromedriver')
driver.maximize_window()
driver.get('https://www.opinet.co.kr/user/main/mainView.do')
time.sleep(1)
driver.find_element_by_css_selector('.ic_m1').click()
time.sleep(1)
driver.find_element_by_css_selector('#SIDO_NM0').send_keys('서울')
time.sleep(1)
driver.find_element_by_css_selector('#SIGUNGU_NM0').send_keys('영등포구')
time.sleep(1)
driver.find_element_by_css_selector('#SIGUNGU_NM0').send_keys('영등포구')
time.sleep(1)
driver.find_element_by_css_selector('#searRgSelect').click()
time.sleep(1)
station_table = driver.find_element_by_css_selector('#body1')
time.sleep(1)
trs = station_table.find_elements_by_tag_name('tr')
name_list = []; price_list = []
for tr in trs:
    tds = tr.find_elements_by_tag_name('td')
    try:
        name = tds[0].find_element_by_tag_name('a').text
        price = tds[1].text
    except:
        name = '?'
        price = '-'
    name_list.append(name)
    price_list.append(price)
driver.quit()
```

```python
df2 = pd.DataFrame({
    'name' : name_list,
    'price' : price_list
})
df2
```

### 도로명 주소 찾기

```python
key_fd = open('../06.RestAPI/roadapikey.txt', mode='r')
roadapi_key = key_fd.read(100)
key_fd.close()
```

```python
roadAddr_list = []
road_url = 'https://www.juso.go.kr/addrlink/addrLinkApi.do'
for name in df2.name.to_list():    
    try : 
        url = f'{road_url}?confmKey={roadapi_key}&currentPage=1&countPerPage=10&keyword={name}&resultType=json'
        urlparse(url).geturl()
        results = requests.get(urlparse(url).geturl()).json()
        addr_list = results['results']['juso']
        addr = addr_list[0]
        roadAddr_list.append(addr['roadAddr'])
    except :
        roadAddr_list.append('-')
```
